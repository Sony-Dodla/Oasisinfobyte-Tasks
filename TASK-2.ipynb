```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Unemployment in India.csv")
```

```python
print(df.head())
print(df.info())
df.columns=df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', errors='coerce')
plt.figure(figsize=(12,6))
sns.lineplot(data=df,x='Date',y="Estimated Unemployment Rate (%)",hue="Region",marker="o")
plt.title("Unemployment Rate over time in india")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(10,5))
region_avg=df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)
sns.barplot(x=region_avg.values,y=region_avg.index,palette="coolwarm")
plt.title("Average Unemployment Rate by Region")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```
